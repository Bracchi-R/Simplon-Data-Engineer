```python
import numpy as np
import pandas as pd
```

```python
url='https://studio.ironhack.school/asset-v1:IRONHACK+DAFT+202006_PAR+type@asset+block@datasets_674388_1186156_tv_shows.csv'
df = pd.read_csv(url, index_col = ('Unnamed: 0'))
```

```python
df.aggregate(['count', min])
```

```python
#find only one value
df.agg({
    'Title' : ['shape', 'max'],
    'Year' : ['median', min],
    'type' : ['median', 'nunique']
}).fillna('').T
```

```python
df.groupby('Netflix').describe().T
```

```python
df[['IMDb', 'Netflix']].groupby('Netflix').mean()
```

```python
df[['IMDb', 'Netflix']].groupby('Netflix').agg([min, max, 'mean'])
```

```python
df[['IMDb', 'Netflix']].groupby('Netflix').agg(random_index = ('IMDb', lambda x : (x.mean()-x.min())/(x.max()-x.min())))
```

```python
def quantile25(x):
    return x.quantile(0.25)

df2 = df.groupby('Netflix').agg({'Year': [quantile25, 'median', max]})
```

```python
df2.columns = ['q', 'm', 'max']
```

```python
df.groupby('Netflix').agg(
    Year_quantile25 = ('Year', quantile25),
    Year_median = ('Year', 'median'),
    IMDb_max = ('IMDb', max)
)
```

```python
#Filter
#filter the series of specific age categories the ones with avg IMDb above 7.3
```

```python
test  = df.groupby('Age').IMDb.mean() #avg IMDb selection
selection = test[test> 7.3].index #index to get the categorie
df[df.Age.isin(selection)]
```

```python
#filter approche (select only true ) (group of age that follow the condition)
df.groupby('Age').filter(lambda x : x.IMDb.mean()>7.3)
```

```python
def compare(x):
    if isinstance(x.Age, str): #if x.Age is not nan cause nan is float
        return x.IMDb-test[x.Age]
    else:
        return np.nan
```

```python
df.apply(compare, axis = 1)
```

```python
#which serie is the best in each categorie
test = df[['IMDb', 'Age']].groupby('Age')
#test
new_series = pd.Series()

for i ,j in test: #i is the group, j is the dataframe
    new_series = new_series.append(j.IMDb.rank(ascending= False))
new_series.sort_index()
```
